Realicemos el proceso de "forward propagation" para la red simple, de una neurona con función de activación lineal.
La "Forward propagation" es el proceso de calcular la salida de la red neuronal a partir de las entradas y los parámetros del modelo (pesos y bias).
La función de activación es como un "filtro" que decide cómo sale el resultado, el hecho de que sea lineal quiere decir que el resultado final será el mismo que el valor arrojado por la función presente en nuestra neurona.

Para este ejercicio, tomaré un arreglo de los valores que "tomarán las x" de mi modelo, haré pasar dichos valores por una función lineal (neurona) y finalmente los mostraré como la salida de la red neuronal mediante una función de activación lineal.

```python
import numpy as np

x = np.array([2,4,6,8,10])

w = 2 #peso
b = 3 #bias

z = (w*x) + b #función lineal (neurona)

a = z #función de activación (al ser lineal se queda igual)

print("para nuestra entrada: ", x)

print("las predicciones de acuerdo a nuestro modelo son:", a)
```
